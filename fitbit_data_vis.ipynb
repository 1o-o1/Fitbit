```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
ds0=pd.read_csv('fitbitnon.csv')
ds1=pd.read_csv('fitbitAsd.csv')
```

```python
one=np.ones(320)
```

```python
one.shape
```

```python
ds1['output']=one
```

```python
zero=np.zeros(320)
ds0['output']=0
```

```python
ds=pd.concat([ds0,ds1],axis=0)
```

```python
ds.head()
```

```python
ds.describe().transpose()
```

```python
plt.figure(figsize=(10,10))
sns.heatmap(ds.corr('spearman'),annot=True,fmt='.2f')
plt.savefig('fitbit_cor_heat.png')
```

```python
plt.figure(figsize=(6,4))
plt.title('Correlation with output')
ds.corr('spearman')['output'].sort_values(ascending=False)[1:].plot(kind='bar')
plt.savefig('fitbit_cor_out.png')
plt.show()
```

```python
fig, axes = plt.subplots(2,2, figsize=(14,7))
sns.distplot(ds['mets'], ax=axes[0,0])
sns.distplot(ds['calories'], ax=axes[0,1])
sns.distplot(ds['level'], ax=axes[1,0])
sns.distplot(ds['sedentary'], ax=axes[1,1])
plt.savefig('fitbitdist.png')
fig.tight_layout()
```
